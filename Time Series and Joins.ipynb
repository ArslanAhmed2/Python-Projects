### Joining tables

__In Pandas for a horizontal combination we have merge() and join(), whereas for vertical combination we can use concat() and append().__

* Merge   (Horizontal)
* Join    (Horizontal)
* Concat  (Vertical)
* Append  (Vertical)


### merge() is used for combining data on common column names or indices.

In case of no common column names, we can use pandas argument left_on and right_on to explicitly indicate pandas on what key columns we want to merge data frames, rest everything remains similar.

```python
sales_product = pd.merge(aw_sales, aw_products, on='ProductKey',how = 'outer')
sales_product.head()
sales_product.shape
```

```python
sales_product_1 = pd.merge(aw_sales, aw_products, on='ProductKey',how = 'inner')
sales_product.head()
sales_product_1.shape
```

```python
pd.set_option('display.max_columns', None)
#products_sales = pd.merge(aw_products, aw_sales)
products_sales_inner = pd.merge(aw_products, aw_sales, how = 'inner', on='ProductKey')
products_sales_outer = pd.merge(aw_products, aw_sales, how = 'outer', on='ProductKey')
products_sales_left = pd.merge(aw_products, aw_sales, how = 'left', on='ProductKey')
products_sales_right = pd.merge(aw_products, aw_sales, how = 'right', on='ProductKey')

products_sales_left
```

### join() is used for combining data on a key column or an index.

By default, pandas join operation is performed on indexes both data frames have default indexes values, so no need to specify any join key, join will implicitly be performed on indexes. 

Default nature of pandas join is left outer join

```python
aw_customers.join(aw_sales, lsuffix='_customers', rsuffix='_sales')
```

```python
aw_customers.join(aw_sales, on= 'CustomerKey', how = 'inner', lsuffix='_customers', rsuffix='_sales', )
```

###  concat() is used for combining Data Frames across rows or columns

concat is by default a vertical operation. axis = 0

```python
aw_combine = pd.concat([aw_products, aw_sales], axis=1, join='inner', ignore_index=False)
aw_combine.shape
```
